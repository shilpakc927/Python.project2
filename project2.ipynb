```python
import pandas as pd
df=pd.read_csv('CLeaned_Students_Performance.csv')
df
```

```python
df.isna().sum()
```

```python
df['lunch']=df['lunch'].fillna(1)
df['lunch']=df['lunch'].astype(int)
```

```python
df['test_preparation_course']=df['test_preparation_course'].fillna(1)
df['test_preparation_course']=df['test_preparation_course'].astype(int)
```

```python
df['math_score']=df['math_score'].fillna(54)
df['math_score']=df['math_score'].astype(int)
```

```python
df['writing_score']=df['writing_score'].fillna(64)
df['writing_score']=df['writing_score'].astype(int)
```

```python
df.dtypes
```

```python
for x in df.index:
    if df.loc[x,"lunch"]>0:
        df.loc[x,"gender"]='male'
    else:
        df.loc[x,"gender"]='female'
```
